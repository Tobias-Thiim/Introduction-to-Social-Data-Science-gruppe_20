````python
import pandas as pd

def process_weather(df):
    """Function to structure and clean weather data.
    
    Structuring includes renaming the columns and selecting 
    only observations of maximum temperature. 
    Cleaning includes inserting missing decimal, sorting and
    resetting the index.
    
    Args:
        df (str): name of csv-file to load
        
    Returns:
        (pd.DataFrame): processed data
    """
    # loads the data
    df_weather = pd.read_csv(str(fp_data)+'/'+df)

    # structure and clean data using methods chaining
    # note that the original columns now are strings when loading the csv file
    # and not integers as when downloading the data
    df_out = \
        df_weather\
            .rename(columns={'0': 'station', '1': 'datetime', '2': 'obs_type', '3': 'obs_value'})\
            .query("obs_type == 'TMAX'")\
            .assign(obs_value=lambda df: df['obs_value']/10)\
            .sort_values(by=['station', 'datetime'], ascending = [True, False])\
            .reset_index(drop=True)\
            .copy() 

    # area process
    df_out['area'] = df_out['station'].str[0:2]

    # datetime process
    df_out['datetime_dt'] = pd.to_datetime(df_out['datetime'], format = '%Y%m%d')
    df_out['month'] = df_out['datetime_dt'].dt.month
    df_out['year'] = df_out['datetime_dt'].dt.year

    return df_out

df_weather_period = process_weather('weather_data_1870-1875.csv')
df_weather_period
````

````python
def get_colspecs(lines):
    """Extracts colspecs from `ghcnd-stations-column-metadata.txt`.
    
    Args:
        lines (list[str]): 
            list of relevant rows from `ghcnd-stations-column-metadata.txt` 
    
    Returns:
        (list[str]): 
            list of extracted colspecs i.e. ['1-11', '13-20', ..., '81-85']
    """
    colspec_idx_start = 13
    colspec_idx_end = 17 + 1  # Including idx 17
    
    colspecs = [line[colspec_idx_start:colspec_idx_end].strip() for line in lines]
    
    return colspecs
    
colspecs = get_colspecs(lines)
colspecs
````

> **Ex. 3.3.3.4:** Write a function named `get_colspec_pair` which takes as input a string variable named `colspec` and returns a tuple of integers. Specifically, the function should take a string similar to each element of `colspecs`, split this string by `-` and return a tuple of integers where
1. The first integer should have `1` subtracted from it (Python is 0-indexed!)
2. The second integer should stay as it is (the intervals provided to the pandas function `read_fwf` should be non-inclusive)
> As an example, applying the function to `"1-11"` and `"13-20"` should yield the following results:

```python
print(get_colspec_pair("1-11"))
## output: (0, 11)

print(get_colspec_pair("13-20"))
## output: (12, 20)
```


````python
def get_conspec_pair(colspec): 
    start_str, end_str = colspec.split('-')
    
    start_index = int(start_str) - 1  # first integer gets 1 subtracted
    end_index = int(end_str)  # second stays the same
    
    return (start_index, end_index)

print(get_conspec_pair("13-20"))
````

````python
import folium

def get_marker(lat, lon, station_name, avg_obs_value, icon='cloud', color="blue"):
    """Creates a `folium.Marker` for a given station
    
    Args:
        (lat): latitude of station
        (lon): longitude of station
        (station_name): name of station
        (avg_obs_value): avg. obs_value for given station
        
    Returns:
        (folium.Marker): object to be added to a folium map
    """
    popup = "\n".join([station_name, f"Avg. obs_value: {avg_obs_value:.2f}"])
    marker = folium.Marker(
        location=[lat, lon],
        popup=popup,
        icon=folium.Icon(icon=icon, color=color)
    )
    return marker

# Create folium map centered on Ontario
COORDS_ONTARIO = (43.40168574192175, -80.33021323830818)
m = folium.Map(location=COORDS_ONTARIO, zoom_start=6)

# Zipper object to iterate through
zipper = zip(
    locations_ontario['LATITUDE'],
    locations_ontario['LONGITUDE'],
    locations_ontario['NAME'],
    locations_ontario['station']
)

# Create markers using a list comprehension
markers_ontario = [
    get_marker(lat, lon, name, avg_obs_value_ontario[station_id])
    for lat, lon, name, station_id in zipper
]

# Add weather station markers to the map
for station_marker in markers_ontario:  
    station_marker.add_to(m)

# Display the map
m
````

````python
# YOUR CODE HERE
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings
from pathlib import Path

warnings.filterwarnings('ignore')




##############   0   ################
titanic = sns.load_dataset('titanic')

titanic.iloc[0:5] # Show the first five rows


#The dataset contains information regarding the passengers on Titanic. The rows represent the indiviudal passengers.
#The columns are variables. For instance there is a variable named "survived" which can take either 0 or 1 depending \\
# on whether the given passener survived or not. Furthermore there is information on age, sex, class etc..


##############   1   ################

font_options = {'family' : 'arial', #Defining standard font options
                'size': 6}
plt.rc('font', **font_options)


plt.figure(figsize=(6, 3), dpi=300) # Set up the figure size


sns.barplot(data=titanic, x='class', y='survived', hue='sex', ci=None,
palette={'male': 'darkred', 'female': 'grey'}) # Create the barplot & choose colors



plt.title('Probability of Survival for Men and Women within Each Passenger Class', fontsize=8) # Add title and labels
plt.xlabel('Passenger Class', fontsize=8)
plt.ylabel('Survival Probability', fontsize=8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False) #removing axis

plt.legend(frameon=False) #removing edges around legend


##############   2   ################
# It doesn’t make sense to use a boxplot to visualize the data. 
#Boxplots show the quantiles and potential outliers.
#Our outcome variable in this case is binary, which means that quantiles and outliers make little sense.  

##############   3   ################
plt.figure(figsize=(6, 3), dpi=300) # Set up the figure size

sns.boxplot(data=titanic, x='class', y='fare', 
            palette={'First': 'darkred', 'Second': 'darkblue', 'Third': 'darkgreen'})


#ax.set_title('Fare Distribution by Passenger Class')

plt.title('Distribution of fare prices', fontsize=8) # Add title and labels
plt.xlabel('Passenger Class', fontsize=8)
plt.ylabel('Fare price', fontsize=8)
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False) #removing axis

##############   4   ################
fp_figs = Path.cwd() / "figs"  
# Use the Path object to actually create the subfolder
Path.mkdir(fp_figs, exist_ok=True) 


##############   5   ################

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5), dpi=300)

# First subplot: Barplot
sns.barplot(data=titanic, x='class', y='survived', hue='sex', ci=None,
            palette={'male': 'darkred', 'female': 'grey'}, ax=axes[0])
axes[0].set_title('Probability of Survival for Men and Women within Each Passenger Class', fontsize=8)
axes[0].set_xlabel('Passenger Class', fontsize=8)
axes[0].set_ylabel('Survival Probability', fontsize=8)
axes[0].spines['top'].set_visible(False)
axes[0].spines['right'].set_visible(False)
axes[0].legend(frameon=False)

# Second subplot: Boxplot
sns.boxplot(data=titanic, x='class', y='fare', 
            palette={'First': 'darkred', 'Second': 'darkblue', 'Third': 'darkgreen'}, ax=axes[1])
axes[1].set_title('Distribution of Fare Prices', fontsize=8)
axes[1].set_xlabel('Passenger Class', fontsize=8)
axes[1].set_ylabel('Fare Price', fontsize=8)
axes[1].spines['top'].set_visible(False)
axes[1].spines['right'].set_visible(False)

# Adjust layout
plt.tight_layout()

# Save the figure
plt.savefig(fp_figs / 'Example2.png')

# Show the plot (optional, can be removed if not needed)
plt.show()
````

> **Ex. 4.3.7:** Use [pairplot with hue](https://seaborn.pydata.org/generated/seaborn.pairplot.html) to create a figure that clearly shows how the different species vary across measurements in the iris dataset. Change the color palette and remove the shading from the density plots. Store the new figure under `figs/pairplot.png`. _Bonus:_ Try to explain how the `diag_kws` argument works (_hint:_ [read here](https://stackoverflow.com/questions/1769403/understanding-kwargs-in-python))

````python
# YOUR CODE HERE
sns.set_context("notebook")

pairplot = sns.pairplot(
    iris,
    hue='species',                  
    palette={'setosa' : 'darkred', 'virginica' : 'darkblue', 'versicolor' : 'darkgreen'},                
    diag_kind='kde',                
    diag_kws={'shade': False}       
)

pairplot.savefig(fp_figs / 'pairplot.png', dpi=300)

# diag_kws is a dictionarie of keyword arguments. It is used to pass arguemnts to the diagonal plots, so it is a way of
# customizing these via a dictionarie.
````
