## DSB stationer
Nu forsøger vi at scrape DSB stationer

```python
url = "https://www.dsb.dk/api/stations/getstationlist"
print(url)
```

```python
import requests # import the module requests
response = requests.get(url) # submit query with `get` and save response as object
```

```python
# Konverterer response til json
response_json = response.json() # convert response to a list of dicts
```

```python
import pandas as pd
pd.DataFrame(response_json)
```
